```python
# ==========================================================
# SECTION 2 : CREATING LISTS
# ==========================================================

"""
OBJECTIVE
---------
Learn every standard way of creating a list.

By the end of this section, you should know:

✓ Empty list
✓ List with values
✓ Mixed datatype list
✓ Nested list
✓ list() constructor

We intentionally avoid advanced concepts here.
Those will come later in the handbook.
"""
```

```python
# ==========================================================
# METHOD 1 : Empty List
# ==========================================================

"""
Syntax:
    variable = []

Purpose:
    Creates an empty list.

When to use:
    Whenever you need a list that will be filled later.
"""

numbers = []

print(numbers)
```

```python
# ==========================================================
# METHOD 2 : List with Initial Values
# ==========================================================

"""
Simply separate values using commas
inside square brackets.
"""

numbers = [10, 20, 30, 40]

print(numbers)
```

```python
cities = ["Delhi", "Mumbai", "Chandigarh"]

print(cities)
```

```python
# ==========================================================
# DUPLICATE VALUES
# ==========================================================

"""
Lists allow duplicate values.
"""

numbers = [5, 5, 5, 10, 10, 15]

print(numbers)
```

```python
# ==========================================================
# HETEROGENEOUS LIST
# ==========================================================

"""
Lists can store different object types together.
"""

employee = [

    101,

    "Vansh",

    True,

    75000.50

]

print(employee)
```

```python
# ==========================================================
# NESTED LIST
# ==========================================================

"""
A list can contain another list.
"""

matrix = [

    [1, 2, 3],

    [4, 5, 6],

    [7, 8, 9]

]

print(matrix)
```

```python
"""Interview Insight ⭐

Nested lists are commonly used to represent:

Matrices
Chess boards
Game boards
Tables
Grids

We'll learn how to access their elements in later sections."""
```

```python
# ==========================================================
# list() CONSTRUCTOR
# ==========================================================

"""
Python provides a built-in constructor:

list()

It creates an empty list when called
without arguments.
"""

numbers = list()

print(numbers)
```
