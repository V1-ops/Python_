Introduction

Traversal means visiting every element of a list, usually one by one.

Almost every DSA algorithm begins with traversal.

Examples include:

Finding the maximum
Computing a sum
Searching
Counting frequencies
Two Pointer problems
Sliding Window
Prefix Sum

Before learning advanced algorithms, you should be comfortable with every way Python lets you traverse a list.

```python
# part 1(for loop)
numbers = [10, 20, 30, 40]

for num in numbers:
    print(num)
```

```python
# Part 2(While loop)
numbers = [10, 20, 30, 40]

i = 0

while i < len(numbers):
    print(numbers[i])
    i += 1
```

When should you use while?

Whenever:

You need to skip elements.
The index changes dynamically.
The stopping condition isn't simply "end of the list."

Examples include:

Two Pointer
Binary Search
Sliding Window

```python
# part 3(range())
numbers = [10, 20, 30]

for i in range(len(numbers)):
    print(i, numbers[i])
```

```python
# Part 4(enumerate())
#Instead of manually tracking indexes:
i = 0

for item in numbers:
    print(i, item)
    i += 1
# use enumerate()
for index, value in enumerate(numbers):
    print(index, value)
```

```python
"""Conceptually:

(0, first_element)

(1, second_element)

(2, third_element)

It doesn't copy the list.

It generates (index, value) pairs lazily.
"""
```

```python
#Part 5 : zip()

#Sometimes you need to traverse multiple lists together.

names = ["Alice", "Bob", "Charlie"]
scores = [90, 80, 95]

for name, score in zip(names, scores):
    print(name, score)
```

```python
numbers = [10, 20, 30]

for num in reversed(numbers):
    print(num)
```

```python
items = ["A", "B", "C", "D", "E", "F"]
# Index:  0    1    2    3    4    5

# Slice from 1 to 4 (requires index 5 as the bound)
for x in reversed(items[1:5]):
    print(x)

# Output:
# E  (index 4)
# D  (index 3)
# C  (index 2)
# B  (index 1)
```

```python
"""Part 7 : Modifying a List While Iterating ⚠️

One of the most common Python mistakes.

Example

numbers = [1, 2, 3, 4]

for num in numbers:
    if num % 2 == 0:
        numbers.remove(num)

print(numbers)

Output

[1, 3]

This appears correct—but only by coincidence.

Consider another example:

numbers = [2, 4, 6, 8]

for num in numbers:
    if num % 2 == 0:
        numbers.remove(num)

print(numbers)

Output

[4, 8]

Why?

Because removing an element shifts the remaining elements left, while the iterator has already advanced to the next index. As a result, some elements are skipped.

Safe Approaches
Option 1 : Iterate over a copy
for num in numbers[:]:
    if num % 2 == 0:
        numbers.remove(num)

Option 2 : Build a new list
result = []

for num in numbers:
    if num % 2 != 0:
        result.append(num)

This is often the cleanest and safest approach.


"""
```

```python
"""🚀 Professional Tips
Use for when you only need values.
Use enumerate() when you need both index and value.
Use zip() when processing related lists together.
Use while when the index changes dynamically.
Never modify a list while iterating over it directly unless you fully understand the consequences."""
```
