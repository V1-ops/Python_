Python provides two ways to sort:

Method	   Modifies Original List?	      Returns New List?
sort()	       ✅ Yes	                        ❌ No
sorted()	    ❌ No	                        ✅ Yes

```python
numbers = [5, 2, 8, 1, 3]

numbers.sort()   #original list is modified and sorted in ascending order

print(numbers)
```

```python
numbers = [3, 1, 2]

result = numbers.sort()

print(result) # returns None because the original list is modified and sorted in ascending order
```

```python
numbers = [5, 2, 8, 1]

new_numbers = sorted(numbers)
# original list is not modified and a new sorted list is returned

print(numbers)

print(new_numbers)
```

Behind the Scenes 🧠

sorted() first creates a new list containing the same references as the original iterable.

Then it sorts that new list.

This is why the original object is unaffected.

```python
numbers = [5, 2, 8, 1]

numbers.sort(reverse=True)

print(numbers)
```

```python
numbers = [5, 2, 8, 1]

print(sorted(numbers, reverse=True))
```

Key parameter


```python
words = ["banana", "kiwi", "apple", "pear"]

words.sort(key=len)

print(words)
```

Internal Working : Timsort 🧠

Python uses an algorithm called Timsort.

Timsort is a hybrid of:

Merge Sort
Insertion Sort

It was designed specifically for real-world data.

Why?

Because real-world data is often partially sorted, not completely random.

Timsort detects these already sorted portions (called runs) and takes advantage of them.

That's why:

Already sorted lists can be processed in O(n) time.
General cases are O(n log n).

This is one of the reasons Python's sorting is extremely fast in practice.
