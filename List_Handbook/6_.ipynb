```python
#Part 1 : remove()

"""Syntax
list.remove(value)

remove() searches for the given value.

Once found,

it removes the first occurrence only.
"""
```

```python
numbers = [10,20,30,20,40]

numbers.remove(20)

print(numbers)

# remove first occurance of 20 from the list.
```

```python
fruits = ["Apple","Banana","Mango"]

fruits.remove("Banana")

print(fruits)
```

2) POP()


```python
numbers = [10,20,30]

value = numbers.pop()

print(value)

print(numbers)
# deletes a number and returns the deleted number 
```

```python
# T.C -> O(n) - worst case, if the value is not present in the list, it will traverse the entire list to find the value.
#O(1)
```


Clear()

```python
numbers = [10,20,30]

numbers.clear()

print(numbers)
# LIST object still exists but it is empty now.
```

Part 4 : del

```python
numbers = [10,20,30]

del numbers[1]

print(numbers)
```

```python
numbers = [1,2,3]

del numbers
# it delets the entire list object. If we try to print it, it will give an error.
```

Choosing the Right Method
Situation	                  Best Choice
Remove known value	            remove()
Remove last element       	    pop()
Remove element at index	        pop(index)
Empty existing list	            clear()
Delete the variable completely	 del

The Interview Answer ⭐

If an interviewer asks:

"Python lists store references. Why is remove() still O(n)?"

A strong answer would be:

"Although a Python list stores references rather than the objects themselves, those references are stored in a contiguous dynamic array. After removing an element, Python must shift all subsequent references one position to the left to maintain contiguity. Each reference assignment is O(1), but shifting up to n references makes the overall operation O(n)."
