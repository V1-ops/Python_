Adding Elements in Python

Python provides three primary ways to add elements.

Method	       Adds     	Position
append()	One object	     End
extend()	Multiple objects	End
insert()	One object	      Anywhere

What is append()?
list.append(item)

Appends exactly one object to the end of the list.

Notice the wording carefully.

One object

Not

One value.

That distinction matters.

```python
numbers = [10,20,30]

numbers.append(40)

print(numbers)
```

```python
fruits = ["Apple","Banana"]

fruits.append("Mango")

print(fruits)
```

```python
numbers = [1,2,3]

numbers.append([4,5])

print(numbers)
```

Visual

Index

0   1   2     3

1   2   3   [4,5]

```python
numbers = [1,2]

result = numbers.append(3)

print(result)
# we cannot assign the result of append() to a variable because it returns None. It modifies the list in place.

print(numbers)
```

Time Complexity

Operation              Complexity,
append()               O(1) Amortized

Part 2 : extend()

```python
#Syntax
#list.extend(iterable)
```

```python
numbers = [1,2,3]

numbers.extend([4,5])

print(numbers)
#unlike append(), extend() takes an iterable and adds each of its elements to the list.
```

Time Complexity
O(k)

where

k

=

number of incoming elements

Part 3 : insert()


```python
# list.insert(index,value)
```

```python
numbers = [10,20,30]

numbers.insert(1,15)

print(numbers)
```

Behind the Scenes 🧠

Python cannot simply "place" the new element.

It has to shift every element after that index.

Time Complexity
Operation	Complexity
insert()	O(n)

```python
# Test Yourself
a = [1,2]

b = [3,4]

a.append(b)

print(a)

c = [1,2]

c.extend(b)

print(c)

d = [10,20,30]

d.insert(1,15)

print(d)
```

Need to add ONE element?
        ↓
     append()

Need to add MANY elements?
        ↓
     extend()

Need to add at a SPECIFIC position?
        ↓
     insert()
