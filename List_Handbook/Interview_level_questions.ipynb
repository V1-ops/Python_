```python
"""1. Flatten Nested Lists ⭐⭐⭐⭐
Problem:-data = [[1, 2], [3, 4, 5], [], [6]]
expected output:-[1, 2, 3, 4, 5, 6]"""

data = [[1, 2], [3, 4, 5], [], [6]]

result = [num for row in data for num in row]

print(result)
# num is at start of the line because the first part of the list comp. is the one which you want to enter into the new list 
# it is read as put the  num to the new list for every row in data and for num in each of the row 
```
