This section covers the list functions and methods that are used most often in real programs and interviews.

Important difference:

- Functions like `len(list)` are called with the list as an argument.
- Methods like `list.append(value)` are called using dot syntax.

## Part 1 : len()

`len()` returns the number of elements in a list.

Time complexity: `O(1)`

```python
numbers = [10, 20, 30, 40]

print(len(numbers))
```

## Part 2 : append()

`append()` adds one element at the end of the list.

Time complexity: usually `O(1)`

```python
fruits = ["apple", "banana"]

fruits.append("mango")

print(fruits)
```

## Part 3 : extend()

`extend()` adds all elements of another iterable to the end of the list.

Use `extend()` when you want to merge values, not insert the entire list as one element.

```python
a = [1, 2]
b = [3, 4]

a.extend(b)

print(a)
```

## Part 4 : insert()

`insert(index, value)` adds a value at a specific position.

Time complexity: `O(n)` because elements may need to shift.

```python
numbers = [10, 20, 40]

numbers.insert(2, 30)

print(numbers)
```

## Part 5 : pop()

`pop()` removes and returns an element.

- `pop()` removes the last element.
- `pop(index)` removes from a specific index.

Last element: `O(1)`

Middle/start element: `O(n)`

```python
numbers = [10, 20, 30, 40]

last = numbers.pop()
second = numbers.pop(1)

print(last)
print(second)
print(numbers)
```

## Part 6 : remove()

`remove(value)` removes the first matching value.

It gives an error if the value is not present.

Time complexity: `O(n)`

```python
names = ["ram", "shyam", "ram", "geeta"]

names.remove("ram")

print(names)
```

## Part 7 : index() and count()

`index(value)` returns the first index of a value.

`count(value)` returns how many times a value appears.

Both are `O(n)`.

```python
marks = [80, 90, 75, 90, 60]

print(marks.index(90))
print(marks.count(90))
```

## Part 8 : sort() and sorted()

`sort()` changes the original list.

`sorted()` returns a new sorted list.

Time complexity: `O(n log n)`

```python
numbers = [40, 10, 30, 20]

new_list = sorted(numbers)

print(numbers)
print(new_list)

numbers.sort()

print(numbers)
```

## Part 9 : reverse() and reversed()

`reverse()` changes the original list.

`reversed()` returns an iterator, so we often convert it using `list()`.

```python
numbers = [1, 2, 3, 4]

reversed_copy = list(reversed(numbers))

print(numbers)
print(reversed_copy)

numbers.reverse()

print(numbers)
```

## Part 10 : min(), max(), sum()

These functions are very useful for numeric lists.

Each one scans the list, so each one takes `O(n)` time.

```python
scores = [70, 85, 90, 60]

print(min(scores))
print(max(scores))
print(sum(scores))
print(sum(scores) / len(scores))
```

## Part 11 : clear(), copy(), list()

`clear()` removes all elements.

`copy()` creates a shallow copy.

`list()` can create a list from another iterable.

```python
original = [1, 2, 3]

duplicate = original.copy()
duplicate.append(4)

print(original)
print(duplicate)

letters = list("abc")
print(letters)

letters.clear()
print(letters)
```

## Interview Insights

- `append()` adds one item, while `extend()` adds many items.
- `remove(value)` removes by value, while `pop(index)` removes by index.
- `sort()` modifies the same list, while `sorted()` creates a new list.
- `reverse()` modifies the same list, while `reversed()` gives an iterator.
- Searching methods like `index()`, `count()`, and `remove()` are `O(n)`.
- Adding/removing at the end is usually fast. Adding/removing near the start is slower because elements shift.
- Many list methods return `None` because they change the list in place.

## Quick Complexity Table

| Operation | Time Complexity |
|---|---|
| `len(lst)` | `O(1)` |
| `lst.append(x)` | Usually `O(1)` |
| `lst.pop()` | `O(1)` |
| `lst.insert(i, x)` | `O(n)` |
| `lst.pop(i)` | `O(n)` |
| `lst.remove(x)` | `O(n)` |
| `x in lst` | `O(n)` |
| `lst.index(x)` | `O(n)` |
| `lst.count(x)` | `O(n)` |
| `lst.sort()` | `O(n log n)` |
