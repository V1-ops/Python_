```python
"""Consider this code:

a = [1, 2, 3]

b = a

b.append(4)

print(a)

Output

[1, 2, 3, 4]

Most beginners expect

[1,2,3]

But that's not what happens.
"""
```

```python
"""What Actually Happens

Python creates one list object.

Both variables point to that same object.

          ┌──────────────┐
a ───────▶│10 │20 │30     │
          └──────────────┘
               ▲
               │
b ─────────────┘

No new list is created.

This is called aliasing.
"""
```

```python
"""Behind the Scenes 🧠

Remember from Section 1:

A list stores references.

Variables also store references.

When Python executes

b = a

it copies only the reference.

Not the list.

This operation is therefore

O(1)

because only one pointer is copied.
"""
```

```python
a = [1,2,3]

b = a

print(id(a))

print(id(b))
```

```python
a = [1,2,3]

b = a

print(a == b)

print(a is b)
```

```python
a = [1,2,3]

b = [1,2,3]

print(a == b)

print(a is b)
```

Creating a real copy


```python
a = [10,20,30]

b = a.copy()

b.append(40)

print(a)

print(b)
```

```python
"""Behind the Scenes

Python creates

a completely new list object.

Then copies all references from the old list.

a ──▶ [10][20][30]

b ──▶ [10][20][30]
"""
```

```python
a = [1,2,3]

b = a[:]

b.append(4)

print(a)

print(b)
#Works same as copy() for one dimensional lists
```

```python
"""Part 4 : Shallow Copy

Here's where things become interesting.

Suppose

a = [[1,2],[3,4]]

Now

b = a.copy()

Looks like

a ──▶ [ RefA ][ RefB ]

b ──▶ [ RefA ][ RefB ]

Notice carefully.

The outer list is copied.

The inner lists are not.

Only their references are copied.

This is called a shallow copy.
"""
```

```python
"""Deep Copy

Sometimes

you want everything copied.

Even nested objects.

Python provides
"""
import copy

a = [[1,2],[3,4]]

b = copy.deepcopy(a)

b[0].append(100)

print(a)

print(b)
```
