```python
# ==========================================================
# WHAT IS INDEXING?
# ==========================================================

"""
Every element inside a list has a unique position.

That position is called its INDEX.

Example:

Index
 0    1    2    3
+----+----+----+----+
| 10 | 20 | 30 | 40 |
+----+----+----+----+

Element 10 is at index 0.
Element 20 is at index 1.
Element 30 is at index 2.
Element 40 is at index 3.

Python uses ZERO-BASED INDEXING.
"""

numbers = [10, 20, 30, 40]

print(numbers)
```

```python

"""
Why does indexing start from 0?
This is one of the most commonly asked interview questions.

Python follows zero-based indexing because it maps naturally
to how arrays are represented in memory.

Conceptually,

Address of element

= Base Address + (Index × Size)

When Index = 0

Address = Base Address

No offset is needed.

Almost every modern programming language
(C, C++, Java, Python, JavaScript)
uses zero-based indexing.
"""
```

```python
# ==========================================================
# POSITIVE INDEXING
# ==========================================================

numbers = [10, 20, 30, 40]

print(numbers[0])
print(numbers[1])
print(numbers[2])
print(numbers[3])
```

```python
student = [

    "Rahul",

    20,

    "Delhi"

]

print(student[0])
print(student[1])
print(student[2])
```

Interview Tip ⭐

Think of the index as the house number, not the person.

House Number → Index

Person → Value

You use the house number (index) to reach the person (value).

```python
# ==========================================================
# MODIFYING AN ELEMENT
# ==========================================================

numbers = [10, 20, 30, 40]

print("Before :", numbers)

numbers[1] = 200

print("After  :", numbers)
```

```python
cities = [

    "Delhi",

    "Mumbai",

    "Pune"

]

cities[2] = "Bangalore"

print(cities)
```

Time Complexity

Updating an element

numbers[2] = 100

takes

O(1)

because Python directly knows where index 2 is.

4. Negative Indexing

Python provides a very convenient feature.

Instead of counting from the beginning,

you can count from the end.

Index

 0    1    2    3

10   20   30   40

-4  -3   -2   -1

```python
# ==========================================================
# NEGATIVE INDEXING
# ==========================================================

numbers = [10, 20, 30, 40]

print(numbers[-1])

print(numbers[-2])

print(numbers[-3])

print(numbers[-4])
```
