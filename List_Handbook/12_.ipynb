List comprehension is a short and clean way to create a new list from an existing iterable.

Basic syntax:

`new_list = [expression for item in iterable]`

Use it when the logic is simple and readable. If the logic becomes confusing, use a normal loop.

```python
# Normal loop
squares = []

for num in range(1, 6):
    squares.append(num * num)

print(squares)

# Same work using list comprehension
squares = [num * num for num in range(1, 6)]

print(squares)
```

## Part 1 : Basic Examples

The first part is the value you want to store in the new list.

`[value_to_store for item in iterable]`

```python
numbers = [1, 2, 3, 4, 5]

doubled = [num * 2 for num in numbers]
cubes = [num ** 3 for num in numbers]
as_strings = [str(num) for num in numbers]

print(doubled)
print(cubes)
print(as_strings)
```

## Part 2 : Filtering With if

Use `if` at the end when you want to keep only selected elements.

Syntax:

`new_list = [expression for item in iterable if condition]`

```python
numbers = [10, 15, 20, 25, 30, 35]

even_numbers = [num for num in numbers if num % 2 == 0]
greater_than_20 = [num for num in numbers if num > 20]

print(even_numbers)
print(greater_than_20)
```

## Part 3 : if-else Inside List Comprehension

Use `if-else` before the `for` when every element must produce an output.

Syntax:

`new_list = [value_if_true if condition else value_if_false for item in iterable]`

```python
marks = [35, 80, 22, 90, 50]

result = ["Pass" if mark >= 40 else "Fail" for mark in marks]

print(result)
```

## Part 4 : Working With Strings

List comprehensions are commonly used to clean or transform text data.

```python
names = ["  ram", "SHYAM  ", "  Geeta  "]

clean_names = [name.strip().title() for name in names]

print(clean_names)
```

## Part 5 : Flattening a 2D List

A nested comprehension can convert a 2D list into a normal 1D list.

Read it from left to right like nested loops.

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

flat = [num for row in matrix for num in row]

print(flat)
```

## Part 6 : Interview Examples

These are small but common interview-style uses.

```python
# 1. Extract vowels from a string
text = "interview"
vowels = [ch for ch in text if ch in "aeiou"]
print(vowels)

# 2. Remove negative numbers
nums = [5, -2, 8, -1, 0]
positive = [num for num in nums if num >= 0]
print(positive)

# 3. Length of each word
words = ["python", "list", "dsa"]
lengths = [len(word) for word in words]
print(lengths)
```

## Time Complexity

For a list of `n` elements, list comprehension usually takes `O(n)` time because it visits each element once.

Space complexity is also usually `O(n)` because a new list is created.

For nested lists with `r` rows and `c` columns, traversal is `O(r * c)`.

## Interview Insights

- List comprehension creates a new list.
- It is usually cleaner than a loop for simple transformations and filters.
- It should not be used when readability becomes poor.
- `[x for x in nums if condition]` filters values.
- `[a if condition else b for x in nums]` transforms every value.
- A generator expression uses parentheses and does not create the full list immediately.
