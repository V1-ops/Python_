Slice Syntax
list[start : stop : step]
Part	Meaning
start	Inclusive
stop	Exclusive
step	Jump size

One of the most important interview rules:

Start is INCLUDED, Stop is EXCLUDED.

```python
numbers = [10,20,30,40,50,60]

print(numbers[1:4])
```

```python
numbers = [10,20,30,40,50]

print(numbers[:3])

#Python assumes the start index to be 0 if not specified. So, numbers[:3] is equivalent to numbers[0:3].
```

```python
numbers = [10,20,30,40,50]

print(numbers[2:])

#Python assumes the stop index to be the length of the list if not specified. So, numbers[2:] is equivalent to numbers[2:len(numbers)].
```

```python
numbers[:]
```

```python
numbers = [10,20,30,40,50,60]

print(numbers[0:6:2])
```

```python
print(numbers[1:5:2])
```

```python
numbers = [10,20,30,40,50]

print(numbers[::-1])
```

```python
word = ["P","Y","T","H","O","N"]

print(word[::-1])
```

```python
numbers = [10,20,30,40,50]

print(numbers[-4:-1])
```

```python
# Copy list using slicing

list = [10,20,30,40]

copied_list = list[:]

print(list)
print(copied_list)
```

🧠 Behind the Scenes

Many beginners think slicing is "free."

It is not.

numbers[2:5]

Python creates a brand new list.

It copies references from the original list into the new list.

Therefore,

numbers[2:5]

is not O(1).

It is

O(k)

where

k = number of elements copied

This is a very common interview question.

Out-of-Range Slicing

Unlike indexing,

Slicing never throws an error.

numbers = [10,20,30]

print(numbers[0:100])

Output

[10,20,30]

Python silently stops at the end.

Another example

print(numbers[100:200])

Output

[]

This is intentional.

🚀 Professional Tip

When solving DSA problems, avoid unnecessary slicing inside loops.

Example:

for i in range(n):
    temp = arr[i:]

This looks innocent.

But every iteration creates a new list, making the overall complexity much worse than expected.

This is a subtle performance issue that experienced Python developers watch for.
