```python
# ==========================================================
# PYTHON DSA HANDBOOK
# Chapter 1 : Lists
# Section 1 : Introduction
# ==========================================================

"""
WHAT IS A LIST?
---------------

A list is Python's built-in dynamic array.

It is:
✓ Ordered
✓ Mutable (can be changed)
✓ Allows duplicate values
✓ Can store different data types
✓ Indexed (starts from 0)

Examples:
---------
[10, 20, 30]

["Python", "Java", "C++"]

[1, "Hello", 3.14, True]

[]
"""

# Example 1
numbers = [10, 20, 30, 40]

print(numbers)

# Example 2
mixed = [10, "Python", 3.14, True]

print(mixed)

# Example 3
empty = []

print(empty)
```

```python
# ==========================================================
# LIST CHARACTERISTICS
# ==========================================================

"""
1. Ordered
---------
Elements keep their insertion order.

2. Mutable
----------
Elements can be modified.

3. Duplicates Allowed
---------------------
Same value can appear multiple times.

4. Heterogeneous
----------------
Different data types can exist together.

5. Dynamic
----------
Python automatically resizes the list when needed.
"""
```

```python
# ==========================================================
# MUTABILITY
# ==========================================================

fruits = ["apple","orange","kiwi"]

print("Before :", fruits)

fruits[1] = "mango"

print("After:", fruits)

```

```python
# ==========================================================
# DUPLICATES
# ==========================================================

numbers = [10, 20, 20, 20, 30, 30]

print(numbers)
```

```python
# ==========================================================
# HETEROGENEOUS LIST
# ==========================================================

employee = [
    101,
    "Vansh",
    85000,
    True,
    ["Python", "SQL", "ML"]
]

print(employee)
```

```python
# ==========================================================
# DYNAMIC ARRAY
# ==========================================================

"""
Unlike C arrays,

Python lists grow automatically.

You DON'T specify size.

Example:

C:
int arr[10];

Python:
arr = []

Keep appending.

Python allocates more memory automatically.
"""

numbers = []

for i in range(10):
    numbers.append(i)

print(numbers)
```

```python
# ==========================================================
# MEMORY MODEL
# ==========================================================

"""
A Python list DOES NOT store the actual objects.

It stores REFERENCES to objects.

Visual Representation

numbers

+----+----+----+
| •  | •  | •  |
+----+----+----+
  |     |     |
  v     v     v
 10    20    30

Each box stores a reference (memory address),
not the integer itself.

This is why lists can store ANY object.
"""

numbers = [10, 20, 30]

print(numbers)
```
